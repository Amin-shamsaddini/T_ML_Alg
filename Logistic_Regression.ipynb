```python
# Import Statements
from pydataset import data
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
#Get Data
titanic = data('titanic')
titanic.sample(10)
```

```python
titanic.iloc[:, 0].values
```

```python
#Feature Engineering (One hot encoding)
titanic= pd.get_dummies(titanic , drop_first=True)
titanic.sample(10)
```

```python
#Train test split
X_train , X_test ,y_train , y_test = train_test_split(titanic.drop('survived_yes' , axis = 1 ) , titanic["survived_yes"])
```

```python
#Train the model using the training data
solver='lbfgs'
LogReg = LogisticRegression(solver=solver)
LogReg.fit(X_train, y_train) #train phase
```

```python
#Predicting if a class-l child-age female survived
LogReg.predict(np.array([[0, 0, 1, 1]]))[0]
```

```python
LogReg.score(X_test, y_test)
```
